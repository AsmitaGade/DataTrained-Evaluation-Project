```python
# Model Building: Linear Regresion
```

```python
#first we will try using Linear regression to predict the Average price of avacados
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.20, random_state=42)
```

```python
x=df.iloc[:,1:]

y=df.iloc[:,0]


print('x train shape', x_train.shape)
print('y train shape', y_train.shape)

print('x test shape', x_test.shape)
print('y test shape', y_test.shape)
```

```python
# Creating and training Linear Regression model
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test,pred))
print('MSE:', metrics.mean_squared_error(y_test,pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,pred)))
```

```python
# lets plot it

plt.scatter(x=y_test, y=pred)
```

```python
# observation: RMSE is low so we can say it is good model
# we can also see from the plot that it is not exactlt a straight line. so we can check decison tree model also to be sore
```

```python
from sklearn.tree import DecisionTreeRegressor
dtc=DecisionTreeRegressor()
dtc.fit(x_train,y_train)
pred=dtc.predict(x_test)
```

```python
plt.scatter(x=y_test, y=pred)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
# Observation: Here we can see from the plot that we are getting somewhat straihjt line
```

```python
print('MAE:', metrics.mean_absolute_error(y_test,pred))
print('MSE:', metrics.mean_squared_error(y_test,pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,pred)))
```

```python
# observation : here we are geeting even low rmse from linear which is good fro our model, so now we have to try to imrove predictions using random forest regressor
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
pred=rfr.predict(x_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test,pred))
print('MSE:', metrics.mean_squared_error(y_test,pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,pred)))
```

```python
# Observation: as we can see here we are getting low rmse from previous two models.
# so we can go with the Randomforestregressor.
```

```python
sns.distplot((y_test-pred), bins=50)
```
