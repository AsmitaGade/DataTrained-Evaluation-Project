# Separate the data
As we have already removed some irrelevant column let us proceed with splitiing the dataset into x and y

```python
y = ibm_df.iloc[:, 1]
x = ibm_df
x.drop('Attrition', axis = 1, inplace = True)
```

```python
x
```

```python
y
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=.25, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
import sklearn

from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
```

```python
dtc=DecisionTreeClassifier()
svc=SVC()
mnb=MultinomialNB()
knn=KNeighborsClassifier(n_neighbors=23)



model=[dtc,mnb,knn,svc]

for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('Accuracy score of ', m, 'is:')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('\n')
```

```python
predm 
```

```python
from sklearn.ensemble import AdaBoostClassifier
ad=AdaBoostClassifier(n_estimators=50)
ad.fit(x_train,y_train)
ad_pred=ad.predict(x_test)
print(accuracy_score(y_test,ad_pred))
print(confusion_matrix(y_test,ad_pred))
print(classification_report(y_test,ad_pred))
```

```python
from sklearn.svm import SVC
svc=SVC()

# CREATE aDAbOOST oBJECT

ad=AdaBoostClassifier(n_estimators=50, base_estimator=svc, algorithm='SAMME')
ad.fit(x_train,y_train)

ad_pred=ad.predict(x_test)
print(accuracy_score(y_test,ad_pred))
print(confusion_matrix(y_test,ad_pred))
print(classification_report(y_test,ad_pred))
```

Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier()

gb.fit(x_train,y_train)
gb_pred=gb.predict(x_test)
print(accuracy_score(y_test,gb_pred))
print(confusion_matrix(y_test,gb_pred))
print(classification_report(y_test,gb_pred))
```
